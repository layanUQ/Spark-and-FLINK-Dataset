```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load all three DataFrames
df1 = pd.read_csv(
    r"D:\arabic_ccnews_partial (1).csv",
    engine='python',
    on_bad_lines='warn'  # Print a warning for each bad line
)

df2 = pd.read_csv(
    r"D:\arabic_ccnews_partial (2).csv",
    engine='python',
    on_bad_lines='warn'  # Print a warning for each bad line
)

df3 = pd.read_csv(
    r"D:\arabic_ccnews_partial.csv",
    engine='python',
    on_bad_lines='warn'  # Print a warning for each bad line
)

# Concatenate the three DataFrames
sample_df = pd.concat([df1, df2, df3], ignore_index=True)

# Check the DataFrame columns
print(sample_df.columns)

# Assuming the publisher is the 12th column and the publication date is the 3rd
publisher_column_name = sample_df.columns[12]  # Change this index if necessary
date_column_name = sample_df.columns[2]

# Convert the publication date column to datetime
sample_df[date_column_name] = pd.to_datetime(sample_df[date_column_name], errors='coerce')

# Extract year from the date
sample_df['year'] = sample_df[date_column_name].dt.year

# Group by publisher and year, then count articles
articles_per_publisher = sample_df.groupby(['year', publisher_column_name]).size().reset_index(name='article_count')

# Find the publisher with the highest number of articles each year
highest_articles_per_year = articles_per_publisher.loc[articles_per_publisher.groupby('year')['article_count'].idxmax()]

# Plot the results
plt.figure(figsize=(12, 6))
sns.barplot(data=highest_articles_per_year, x='year', y='article_count', hue=publisher_column_name)
plt.title('Publisher with Highest Number of Articles Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.legend(title=publisher_column_name)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load all three DataFrames
df1 = pd.read_csv(
    r"D:\arabic_ccnews_partial (1).csv",
    engine='python',
    on_bad_lines='warn'  # Print a warning for each bad line
)

df2 = pd.read_csv(
    r"D:\arabic_ccnews_partial (2).csv",
    engine='python',
    on_bad_lines='warn'  # Print a warning for each bad line
)

df3 = pd.read_csv(
    r"D:\arabic_ccnews_partial.csv",
    engine='python',
    on_bad_lines='warn'  # Print a warning for each bad line
)

# Concatenate the three DataFrames
sample_df = pd.concat([df1, df2, df3], ignore_index=True)

# Check the DataFrame columns
print(sample_df.columns)

# Assuming the publication date is in the 3rd column
date_column_name = sample_df.columns[2]  # Change this index if necessary

# Convert the publication date column to datetime
sample_df[date_column_name] = pd.to_datetime(sample_df[date_column_name], errors='coerce')

# Extract the month from the date
sample_df['month'] = sample_df[date_column_name].dt.month

# Count the occurrences of each month
month_counts = sample_df['month'].value_counts().reset_index()
month_counts.columns = ['Month', 'Count']

# Determine the most common month
most_common_month = month_counts.loc[month_counts['Count'].idxmax()]

# Plotting the results
plt.figure(figsize=(12, 6))
sns.barplot(data=month_counts, x='Month', y='Count', palette='viridis')
plt.title('Number of Articles Published by Month')
plt.xlabel('Month')
plt.ylabel('Number of Articles')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.axhline(most_common_month['Count'], color='red', linestyle='--', label='Most Common Month Count')
plt.legend()
plt.show()

print(f"The most common publishing month is: Month {most_common_month['Month']} with {most_common_month['Count']} articles.")
```
